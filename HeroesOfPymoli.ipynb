```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import csv
import os
# File to Load (Remember to Change These)
csvpath = os.path.join('purchase_data.csv')
file_to_load = "purchase_data.csv"
#pd.options.display.float_format = '${:,.2f}'.format
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
all_purchases = len(purchase_data)
totalplayers = purchase_data["SN"].nunique(dropna=True)
#test = purchase_data.groupby(['Gender', "SN"])["Price"].sum().reset_index()
df1 = purchase_data.groupby(['Gender','SN']).sum()['Price'].reset_index()
nummale = len(purchase_data[purchase_data['Gender'] == 'Male'])
numfemale = len(purchase_data[purchase_data['Gender'] == 'Female'])
numother = len(purchase_data[purchase_data['Gender'] == 'Other / Non-Disclosed'])
ud = df1['Gender'].value_counts()
ud1 = pd.DataFrame(ud).reset_index()
ud1.columns = ['gender', 'count']
um = 484
uf = 81
uo = 11
for i, row in ud1.iterrows():
    if row[1] == "Male":
        um == row[2]
    elif row[1] == "Female":
        uf == row[2]  
    elif row[1] == "Other / Non-Disclosed":
        uo == row[2]
femspending = []
malespending = []
otherspending = []
for i, row in df1.iterrows():
    if i>=0 and i<numfemale:
        femspending.append(row[2])
    elif i>=numfemale and i<nummale:
        malespending.append(row[2])
    elif i>=nummale:
        otherspending.append[row[2]]
tp = {"Total Players":totalplayers}
tpf = pd.DataFrame(tp, index=[0])
tpf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#print(str(nummale) + "\n" + str(numfemale)+"\n" + str(numother))
genderdem = pd.DataFrame(columns = ["Total Count", "Percentage of Players"])
maleperc = nummale/all_purchases
femperc = numfemale/all_purchases
otherperc = numother/all_purchases
#print(maleperc)
genderdem = genderdem.append([{"Total Count":nummale, "Percentage of Players":"%.00f%%" % (maleperc * 100)}])
genderdem = genderdem.append([{"Total Count":numfemale, "Percentage of Players":"%.00f%%" % (femperc * 100)}])
genderdem = genderdem.append([{"Total Count":numother, "Percentage of Players":"%.00f%%" % (otherperc * 100)}])
genderdem.reset_index(drop=True, inplace=True)
genderdem.rename(index={0:'Male',1:'Female',2:"Other/Nondisclosed"}, inplace = True)
genderdem
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genderanalysis = pd.DataFrame(columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total Purchase per Person"])
malecash = 0
femalecash = 0 
othercash = 0
with open(csvpath, newline='', encoding="utf8") as csvfile:    
    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    for row in csvreader:
        if str(row[3]) == "Male":
            malecash = malecash + float(row[6])
        elif str(row[3]) == "Female":
            femalecash = femalecash + float(row[6])
        elif str(row[3]) == "Other / Non-Disclosed":
            othercash = othercash + float(row[6])
#print(str(malecash) + "\n" + str(femalecash)+"\n" + str(othercash))
genderanalysis = genderanalysis.append([{"Purchase Count":numfemale, "Average Purchase Price":"$%.2f" % (femalecash/numfemale), "Total Purchase Value":"$%.2f" % femalecash, "Average Total Purchase per Person":"$%.2f" % (femalecash/uf) }])
genderanalysis = genderanalysis.append([{"Purchase Count":nummale, "Average Purchase Price":"$%.2f" %(malecash/nummale), "Total Purchase Value": "$%.2f" % malecash, "Average Total Purchase per Person":"$%.2f" % (malecash/um)}])
genderanalysis = genderanalysis.append([{"Purchase Count":numother, "Average Purchase Price":"$%.2f" % (othercash/numother), "Total Purchase Value": "$%.2f" % othercash, "Average Total Purchase per Person": "$%.2f" % (othercash/uo)}])
genderanalysis.reset_index(drop=True, inplace=True)
genderanalysis.rename(index={0:'Female',1:'Male',2:"Other/Nondisclosed"}, inplace = True)
genderanalysis
```
